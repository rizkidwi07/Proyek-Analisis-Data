### RFM Analysis

```python
orders_df["order_approved_at"] = pd.to_datetime(orders_df["order_approved_at"], errors='coerce')

rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_approved_at": "max",  # mengambil tanggal order terakhir
    "order_id": "nunique",  # menghitung jumlah order
    "price": "sum"  # menghitung jumlah harga yang dihasilkan
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_approved_at"].dropna().dt.date.max()

min_order_date = rfm_df["max_order_timestamp"].dropna().min()
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].fillna(min_order_date)

rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
# Create a 3-row and 1-column layout
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(10, 18))  # Adjust nrows to 3

# Define color palette
colors = ["#72BCD4", "#72BCD4", "#72BCD4"]

# Recency plot with top 3 customers
sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(2), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=12)

# Frequency plot with top 3 customers
sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(2), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=12)

# Monetary plot with top 3 customers
sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(2), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=12)

# Adjust the overall title
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to fit the title
plt.show()
```

**Insight:**
- Dari visualisasi data di atas, dapat dilihat beberapa pelanggan terbaik berdasarkan ketiga parameter RFM tersebut.

## Conclusion

- **Conclution pertanyaan 1**

- **Conclution pertanyaan 2**

Dengan demikian, kota ini menjadi fokus utama dalam pemasaran, promosi, dan distribusi. Selain itu, ini bisa menjadi peluang bagi perusahaan untuk memperluas penetrasi pasar ke kota-kota lain di Brasil.
- **Conclution RFM Analysis (pertanyaan 3, 4, 5)**
  
  Berdasarkan analisis RFM (Recency, Frequency, Monetary), beberapa pelanggan dapat diidentifikasi sebagai pelanggan terbaik. Pelanggan yang memiliki nilai recency rendah adalah mereka yang baru-baru ini melakukan transaksi, pelanggan dengan frequency tinggi menunjukkan loyalitas karena sering berbelanja, dan pelanggan dengan monetary tinggi menunjukkan mereka menghabiskan banyak uang untuk transaksi di e-commerce tersebut. Ini mengindikasikan bahwa perusahaan dapat memprioritaskan pelanggan ini dalam strategi retention atau loyalitas, karena mereka adalah pelanggan yang berharga bagi bisnis.
